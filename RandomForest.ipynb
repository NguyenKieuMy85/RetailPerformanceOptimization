```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('SupplyChainSalesDatasets.csv')

# Display the first few rows of the dataframe
df.head()
```

```python
# Drop the Unnamed: 28 column
df = df.drop('Unnamed: 28', axis=1)

# Drop rows with missing values (NaN)
df = df.dropna()

# Display info about the cleaned dataset
print("\nDataset info after cleaning:")
print(df.info())

# Display the number of rows before and after cleaning
print("\nNumber of rows in cleaned dataset:", len(df))

# Show the first few rows of cleaned data
df.head()

```

```python
# Lọc ra những đơn hàng có Returned = 'Yes'
returned_orders = df[df['Returned'] == 1]  # Assuming 'Returned' is encoded as 1 for 'Yes'

# Lấy ra các đơn hàng bị trả lại
returned_orders_list = returned_orders.index.tolist()  # Using the index as a unique identifier
# Hiển thị các đơn hàng bị trả lại
print("\nCác đơn hàng bị trả lại:")
print(len(returned_orders_list))

# Lấy ra các đơn hàng không bị trả lại
non_returned_orders = df[df['Returned'] == 0]  # Assuming 'Returned' is encoded as 0 for 'No'

# Lấy ra các đơn hàng không bị trả lại
non_returned_orders_list = non_returned_orders.index.tolist()  # Using the index as a unique identifier
# Hiển thị các đơn hàng không bị trả lại
print("\nCác đơn hàng không bị trả lại:")
print(len(non_returned_orders_list))
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Select features for prediction
categorical_features = ['ShipMode', 'Segment', 'Category', 'SubCategory', 'Region']
numerical_features = ['Sales', 'Quantity', 'Discount', 'Profit', 'Cost', 'UnitCP', 'UnitSP', 'Days']

# Create feature matrix X and target vector y
X = df[categorical_features + numerical_features]
y = df['Returned']
```

```python
# Convert categorical variables to numerical using Label Encoding
label_encoders = {}
for feature in categorical_features:
    label_encoders[feature] = LabelEncoder()
    X[feature] = label_encoders[feature].fit_transform(X[feature])
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Print number of training and testing samples
print("\nSố lượng dòng train:", X_train.shape[0])
print("Số lượng dòng test:", X_test.shape[0])
```

```python
# Visualize distribution of returned orders in train and test sets
plt.figure(figsize=(10, 6))

# Count values for training set
train_counts = pd.Series(y_train).value_counts()
test_counts = pd.Series(y_test).value_counts()

# Create bar positions
x = np.arange(2)
width = 0.35

# Create bars
plt.bar(x - width/2, train_counts, width, label='Train')
plt.bar(x + width/2, test_counts, width, label='Test')

# Customize plot
plt.xticks(x, ['Not', 'Returned'])
plt.ylabel('Number of Orders')
plt.title('Distribution of Returned Orders in Train and Test Sets')
plt.legend()

# Add value labels on top of bars
for i, v in enumerate(train_counts):
    plt.text(i - width/2, v, str(v), ha='center', va='bottom')
for i, v in enumerate(test_counts):
    plt.text(i + width/2, v, str(v), ha='center', va='bottom')

plt.show()
```

### Random forest

```python
from imblearn.ensemble import BalancedRandomForestClassifier
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

rf_model = BalancedRandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_res, y_train_res)
y_pred_rf = rf_model.predict(X_test)

# Đánh giá mô hình Random Forest
from sklearn.metrics import  accuracy_score

accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"\nAccuracy Score: {accuracy_rf:.4f}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
import matplotlib.pyplot as plt

# Đọc và làm sạch dữ liệu
df = pd.read_csv("SupplyChainSalesDatasets.csv")
df = df.drop(columns=["Unnamed: 28"], errors="ignore")
df = df.dropna()

# Lựa chọn đặc trưng và mục tiêu
features = ['Sales', 'Profit', 'Discount', 'Quantity', 'Segment', 'Category', 'Region']
target = 'Returned'
df = df[features + [target]]

# Mã hóa các biến phân loại
le_dict = {}
for col in ['Segment', 'Category', 'Region', 'Returned']:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    le_dict[col] = le

# Tách tập huấn luyện và kiểm tra
X = df[features]
y = df[target]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Huấn luyện mô hình Random Forest
rf_model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
rf_model.fit(X_train, y_train)

# Lấy một cây ra từ Random Forest (cây số 0)
estimator = rf_model.estimators_[0]

# Vẽ cây quyết định
fig = plt.figure(figsize=(55, 18), dpi=150)  # tăng figsize và dpi để nét hơn
_ = tree.plot_tree(
    estimator,
    feature_names=features,
    class_names=le_dict['Returned'].classes_,
    filled=True,
    rounded=True,
    fontsize=12
)
plt.title("Một Cây Quyết Định Trong Random Forest", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Re-import libraries and reload the dataset due to code state reset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier

# Load the CSV file again
df = pd.read_csv('SupplyChainSalesDatasets.csv')

# Clean and prepare the dataset
df = df.drop(columns=["Unnamed: 28"], errors='ignore')
df = df.dropna()

# Define features and target
features = ['Sales', 'Profit', 'Discount', 'Quantity', 'Segment', 'Category', 'Region']
target = 'Returned'

# Encode categorical columns
df_model = df[features + [target]].copy()
le_dict = {}
for col in ['Segment', 'Category', 'Region', 'Returned']:
    le = LabelEncoder()
    df_model[col] = le.fit_transform(df_model[col])
    le_dict[col] = le

# Train Random Forest model
X = df_model[features]
y = df_model[target]
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)

# Predict and add the 'predict_return' column
df['predict_return'] = model.predict(X)
df['predict_return'] = le_dict['Returned'].inverse_transform(df['predict_return'])

# Save to a new CSV file
output_path = "D:\\Data\\Sem_2(2024_2025)\\BI\\output.csv"  # Specify a complete file path
df.to_csv(output_path, index=False)

# Visualize Sales vs predict_return
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='predict_return', y='Sales')
plt.title("Phân phối Doanh số (Sales) theo Dự đoán Trả hàng (predict_return)")
plt.xlabel("Dự đoán Trả hàng")
plt.ylabel("Doanh số (Sales)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Biểu đồ phân phối Profit theo dự đoán trả hàng
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='predict_return', y='Profit')
plt.title("Phân phối Lợi nhuận (Profit) theo Dự đoán Trả hàng (predict_return)")
plt.xlabel("Dự đoán Trả hàng")
plt.ylabel("Lợi nhuận (Profit)")
plt.grid(True)
plt.tight_layout()
plt.show()

# Biểu đồ phân phối Cost theo dự đoán trả hàng
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='predict_return', y='Cost')
plt.title("Phân phối Chi phí (Cost) theo Dự đoán Trả hàng (predict_return)")
plt.xlabel("Dự đoán Trả hàng")
plt.ylabel("Chi phí (Cost)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
